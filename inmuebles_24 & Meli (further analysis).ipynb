```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
inmuebles24_df = pd.read_csv('csv_files/COMPLETE_inmuebles_24_CDMX.csv', encoding = 'utf16')
inmuebles24_df
```

```python
dt_string = 'Inmuebles 24'
inmuebles24_df.index = pd.Series([dt_string] * len(inmuebles24_df))
inmuebles24_df = inmuebles24_df.reset_index()
inmuebles24_df = inmuebles24_df.rename(columns={"index":"source_of_info"})
inmuebles24_df.head(2)
```
