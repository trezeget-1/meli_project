# Cleaning of data

```python
final_propiedades_24_df = pd.read_csv('csv_files/COMPLETE_inmuebles_24_CDMX.csv', encoding = 'utf16')
```

```python
final_propiedades_24_df.shape
```

```python
final_propiedades_24_df
```

```python
final_propiedades_24_df['total_area'] = pd.to_numeric(final_propiedades_24_df['total_area'])
```

```python
for x in range(len(final_propiedades_24_df)):
    try:
        if x == 25:
            final_propiedades_24_df['land_size'][x] = int(final_propiedades_24_df['total_area'][x])
        
        final_propiedades_24_df['land_size'][x] = int(final_propiedades_24_df['land_size'][x])
        
    except:
        print(f'index: {x}')
```

```python
final_propiedades_24_df.land_size = pd.to_numeric(final_propiedades_24_df.land_size)
```

```python
for x in range(len(final_propiedades_24_df)):
    if final_propiedades_24_df['Tipo de propiedad'][x] == 'Departamento':
        if final_propiedades_24_df['land_size'][x] == final_propiedades_24_df['total_area'][x]:
            final_propiedades_24_df['CHECK_SIZE'][x] = 'okay'
            
        elif (final_propiedades_24_df['land_size'][x] == 0 and final_propiedades_24_df['total_area'][x] != 0):
            final_propiedades_24_df['CHECK_SIZE'][x] = 'almost_solved'
            final_propiedades_24_df['land_size'][x] = final_propiedades_24_df['total_area'][x]
            
        
        elif (final_propiedades_24_df['total_area'][x] == 0 and final_propiedades_24_df['land_size'][x] != 0):
            final_propiedades_24_df['CHECK_SIZE'][x] = 'almost_solved'            
            final_propiedades_24_df['total_area'][x] = final_propiedades_24_df['land_size'][x]
        
        else:
            final_propiedades_24_df['CHECK_SIZE'][x] = 'to_check'
            
            
#         elif (final_propiedades_24_df['land_size'][x] > final_propiedades_24_df['total_area'][x]):
#             final_propiedades_24_df['CHECK_SIZE'][x] = 'land_size'
```

```python
final_propiedades_24_df['DIFERENCIA'] = ''
final_propiedades_24_df['BEST_DECISION'] = ''
```

```python
appartments_df = final_propiedades_24_df.loc[final_propiedades_24_df['Tipo de propiedad'] == 'Departamento' , :]
appartments_df.head(5)
```

```python
appartments_df.loc[appartments_df['CHECK_SIZE'] == 'almost_solved' , :]
```

```python
almost_solved_app_df = appartments_df.loc[appartments_df['CHECK_SIZE'] == 'to_check' , :]
almost_solved_app_df = almost_solved_app_df.reset_index()
almost_solved_app_df
```

```python
for x in range(len(almost_solved_app_df)):
    
    if almost_solved_app_df['land_size'][x] > almost_solved_app_df['total_area'][x]:
        almost_solved_app_df['DIFERENCIA'][x] = ((almost_solved_app_df['land_size'][x] - almost_solved_app_df['total_area'][x]) / almost_solved_app_df['total_area'][x])*100
        almost_solved_app_df['BEST_DECISION'][x] = 'land_size'
        
#         if almost_solved_app_df['DIFERENCIA'][x] < 50:
#             index_to_edit = almost_solved_app_df['index'][x]
#             print(index_to_edit)
            
            
#             final_propiedades_24_df['land_size'][index_to_edit] = almost_solved_app_df['total_area'][x]

    elif (almost_solved_app_df['total_area'][x] > almost_solved_app_df['land_size'][x]):
        almost_solved_app_df['DIFERENCIA'][x] = ((almost_solved_app_df['total_area'][x] - almost_solved_app_df['land_size'][x]) / almost_solved_app_df['land_size'][x])*100
        almost_solved_app_df['BEST_DECISION'][x] = 'total_area'
#         if almost_solved_app_df['DIFERENCIA'][x] < 50:
#             index_to_edit = almost_solved_app_df['index'][x]
#             print(index_to_edit)
            
            
#             final_propiedades_24_df['land_size'][index_to_edit] = almost_solved_app_df['total_area'][x]
```

### For the houses properties, I set the property size are as the biggest one between land size and total area

### Delete specific results

```python
url_to_delete = 'https://www.inmuebles24.com/propiedades/casa-en-lomas-altas-miguel-hidalgo-ciudad-de-mexico-60708333.html'
```

```python
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final['Link de la publicación'] == url_to_delete].index)

tabla_final = tabla_final.reset_index()

del tabla_final['index']
```

### Filter to apply

```python
tabla_final = tabla_final.drop(tabla_final.loc[(tabla_final['Número de m2']>=800) & (tabla_final['Tipo de propiedad'] == 'Departamento')].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final["Número de baños completos"] == 'not_available' , :].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final["Número de recámaras"] == 'not_available' , :].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final['Número de baños completos']<1,:].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final['Número de recámaras']<1,:].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final['Número de baños completos']>10,:].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final['Longitud'].isna() == True , :].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final['Latitud'].isna() == True , :].index)
tabla_final = tabla_final.drop(tabla_final.loc[ tabla_final['Latitud'] == 'hidden_address' ].index)
tabla_final = tabla_final.drop(tabla_final.loc[ tabla_final['Longitud'] == 'hidden_address' ].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final['Precio'] <= 500000].index)
tabla_final = tabla_final.drop(tabla_final.loc[tabla_final['Precio']>=700000000,:].index)
tabla_final = tabla_final.drop(tabla_final.loc[(tabla_final['Número de m2']>=800) & (tabla_final['Precio'] < 2000000)].index)
```
