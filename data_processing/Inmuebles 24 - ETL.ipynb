### For the houses properties, I set the property size are as the biggest one between land size and total area

```python
for x in range(len(final_propiedades_24_df)):
    
    if final_propiedades_24_df['Tipo de propiedad'][x] == 'Casa':
        
        land_size = final_propiedades_24_df['land_size'][x]        
        total_area = final_propiedades_24_df['total_area'][x]
        
        if land_size > total_area: 
            final_propiedades_24_df['total_area'][x] = land_size
            print(x)
        elif total_area > land_size: 
            final_propiedades_24_df['land_size'][x] = total_area
        
```

```python
# Check if the change was applied and that the measures "land size" and "total area" are the same

for x in range(len(final_propiedades_24_df)):
    
    if final_propiedades_24_df['Tipo de propiedad'][x] == 'Casa':
        
        land_size = final_propiedades_24_df['land_size'][x]        
        total_area = final_propiedades_24_df['total_area'][x]
        
        if land_size != total_area: 
            final_propiedades_24_df['total_area'][x] = land_size
            print(x)
        
```
